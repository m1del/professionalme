# LinkedIn Headshot Exploratory Data Analysis

See more at: https://www.kaggle.com/datasets/manishkumar7432698/linkedinuserprofiles

```python
import os
import cv2
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import requests
from io import BytesIO
from PIL import Image

import random
```

## Data Loading

```python
dataset_path = '../data/raw/LinkedIn/LinkedIn people profiles datasets.csv'

df = pd.read_csv(dataset_path)
image_files = []

for index, row in df.iterrows():
    if len(image_files) >= 25:
        break
        
    if row['avatar'] == "https://static.licdn.com/sc/h/244xhbkr7g40x6bsu4gi6q4ry" or row['avatar'] == "null":
        continue
    
    try:
        response = requests.get(row['avatar'])
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))
        
        image_files.append(img)
    except requests.exceptions.HTTPError as http_err:
        pass
    except Exception as err:
        pass
```

```python
# Load and display 5 sample images
sample_images = random.sample(image_files, 5)

for img in sample_images:
    plt.imshow(img)
    plt.show()
```
